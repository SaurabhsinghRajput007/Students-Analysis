```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info ()
```

```python
df  = df.drop("Unnamed: 0",  axis = 1)
print(df.head())
```

```python
df.head()
```

```python
plt.figure(figsize=  (5,5))
ax= sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
#number of female is more than number of male
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean' , "ReadingScore" : "mean" ,"WritingScore": "mean"})
print(gb)
 
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
#This chart shows education of parents have major impact on grades
```

```python
gb = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean' , "ReadingScore" : "mean" ,"WritingScore": "mean"})
print(gb)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb, annot = True)
plt.show()
#no impact of grades due to maritial status
```
